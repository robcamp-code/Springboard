```python
# First, import the relevant modules
import collections
import requests
```

```python
# Now, call the Quandl API and pull out a small sample of the data (only one day) to get a glimpse
# into the JSON structure that will be returned

params = {
    "start_date" : "2017-01-01",
    "end_date" : "2018-01-01",
    "api_key": API_KEY
}
req = f"https://www.quandl.com/api/v3/datasets/FSE/AFX_X/data.json"

response = requests.get(req, params=params)
```

```python
# Inspect the JSON structure of the object you created, and take note of how nested it is,
# as well as the overall structure
print(type(response))
print(type(response.json()))
```

```python
response_dict = response.json()
```

```python
response_dict['dataset_data']['data']
#response_dict['dataset_data']['column_names']
```

```python
Day = collections.namedtuple(
    'Day', 
    'Open High Low Close Change Traded_Volume Turnover Last_Price Daily_Units Daily_Turnover'
)
```

```python
dictionary = collections.OrderedDict()
for d, o, h, l, c, ch, tv, t, lp, du, dt in response_dict['dataset_data']['data']:
    day = Day(Open=o, High=h, Low=l,
              Close=c,Change=ch,Traded_Volume=tv, 
              Turnover=t, Last_Price=lp,
              Daily_Units=du, Daily_Turnover=dt)
    dictionary[d] = day
dictionary   
```

### Calculate what the highest and lowest opening prices were for the stock in this period.

```python
opening_high = 0
opening_low = 10000000
for v in dictionary.values():
    if v.Open:
        if v.Open > opening_high:
            opening_high = v.Open
        if v.Open < opening_low:
            opening_low = v.Open
    
```

```python
print(opening_high)
print(opening_low)
```

### What was the largest change in any one day (based on High and Low price)?

```python
largest_change = 0
for v in dictionary.values():
    if v.Low and v.High:
        change = v.High - v.Low
        if change > largest_change:
            largest_change = change
largest_change        
```

### What was the largest change between any two days (based on Closing Price)?

```python
reversed_dict = dict(reversed(dictionary.items()))
```

```python
previous = None
greatest_change = 0
for k,v in reversed_dict.items():
    if v.Close:
        
        if previous:
            change = v.Close - previous
            if change > greatest_change:
                greatest_change = change
            previous = v.Close
            
        else: 
            previous = v.Close
            continue
    else:
        previous = None
greatest_change  
```

### What was the average daily trading volume during this year?

```python
avg_traded_volumme = 0
total_traded_volume = 0
number_of_days=0
for v in dictionary.values():
    if v.Traded_Volume:
        number_of_days += 1
        total_traded_volume += v.Traded_Volume
avg_traded_volume = total_traded_volume / number_of_days
avg_traded_volume
```
